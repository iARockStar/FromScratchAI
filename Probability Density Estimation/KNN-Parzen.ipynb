## 1-1 ، پنجره پارزن

```python
def parzen_window(x_train, y_train, x_test, hn):
    predictions = []
    unique_classes = np.unique(y_train)
      
    for x in x_test:
        densities = []
        for current_class in unique_classes:
            thisclassdata = x_train[y_train == current_class]
            N = len(thisclassdata)
            d = x_train.shape[1]
            V = hn**d
            ni = 0
            for thisx in thisclassdata:
                in_window = 1
                for i in range(len(x)):
                    if abs((x[i] - thisx[i]) / hn) > 0.5:
                        in_window = 0
                        break
                ni += in_window

            density = ni / (N * V)
            densities.append(density)

        predictions.append(unique_classes[np.argmax(densities)])

    return np.array(predictions)
```

```python
best_hn = None
best_accuracy = 0
hn_values = []
accuracies = []
for h in range(2, 100):
    y_pred = parzen_window(x_train_iris, y_train_iris, x_test_iris, h)  
    accuracy = np.mean(y_pred == y_test_iris)
    hn_values.append(h)
    accuracies.append(accuracy)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_hn = h
print("best hn =", best_hn, "best accuracy =", best_accuracy)

plt.figure(figsize=(8, 6))
plt.plot(hn_values, accuracies, marker='o')
plt.title("effect of hn parameter on accuracy in iris dataset")
plt.xlabel("hn")
plt.ylabel("accuracy")
plt.grid()
plt.show()
```

## بخش 2-2 ، KNN

```python
def KNN(x_train, y_train, x_test, k):
    result = []

    for x in x_test:
        distances = []
        
        for thisx in x_train:
            dist = np.sqrt(np.sum((x - thisx) ** 2))
            distances.append(dist)
        
        nearest = []
        for _ in range(k):
            min_distance = float('inf')
            min_index = -1
            for i in range(len(distances)):
                if distances[i] < min_distance and distances[i] != 0 and distances[i] != float('inf'):  
                    min_distance = distances[i]
                    min_index = i
            nearest.append(y_train[min_index])
            distances[min_index] = float('inf')
 
        counts = {}
        for i in nearest:
            if i in counts:
                counts[i] += 1
            else:
                counts[i] = 1
      
        max_label = max(counts, key=counts.get)
   
        result.append(max_label)
    
    return np.array(result)
```

```python
best_k = None
best_accuracy = 0
k_values = []
accuracies = []
for k in range(2, 100):
    y_pred = KNN(x_train_iris, y_train_iris, x_test_iris, k)  
    accuracy = np.mean(y_pred == y_test_iris)
    k_values.append(k)
    accuracies.append(accuracy)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_k = k
print("best k =", best_k, "best accuracy =", best_accuracy)

plt.figure(figsize=(8, 6))
plt.plot(k_values, accuracies, marker='o')
plt.title("effect of K parameter on accuracy in iris dataset")
plt.xlabel("K")
plt.ylabel("accuracy")
plt.grid()
plt.show()
```

## بخش دوم

```python
def parzen_knn(x_train, y_train, x_test, k, h):
    results = []
    for x in x_test:
        votes_count = {0: 0, 1: 0, 2: 0}
        feature_count = x_train.shape[1]
        
        for ith_feature in range(feature_count):
            knn_predictions = KNN(x_train[:, ith_feature].reshape(-1, 1), y_train, x[ith_feature].reshape(-1, 1), k)
            for pred in knn_predictions:
                votes_count[pred] += 0.125

            parzen_predictions = parzen_window(x_train[:, ith_feature].reshape(-1, 1), y_train, x[ith_feature].reshape(-1, 1), h)
            for pred in parzen_predictions:
                votes_count[pred] += 0.125

        best_class = max(votes_count, key=votes_count.get)
        results.append(best_class)
    
    return np.array(results)

all_k = [5, 10, 20, 35, 40]
all_h = [0.1, 1, 2, 5, 6]

accuracies = []
for k in all_k:
    for h in all_h:
        y_pred = parzen_knn(x_train_iris, y_train_iris, x_test_iris, k, h)
        accuracy = np.mean(y_pred == y_test)
        accuracies.append((k, h, accuracy))

accuracies_2d = []
for k in all_k:
    row = []
    for h in all_h:
        found = False
        for record in accuracies:
            if record[0] == k and record[1] == h:
                row.append(record[2])
                found = True
                break     
    accuracies_2d.append(row)

accuracies = np.array(accuracies_2d)
print(f"Accuracies shape: {accuracies.shape}")
plt.figure(figsize=(10, 6))
for i, k in enumerate(all_k):
    plt.plot(all_h, accuracies[i, :], marker='o', label=f'k={k}')
plt.xlabel("hn")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()


best_accuracy = -float('inf')  
best_k = None
best_h = None

for i in range(len(all_k)):
    for j in range(len(all_h)):
        if accuracies[i, j] > best_accuracy:
            best_accuracy = accuracies[i, j]
            best_k = all_k[i]
            best_h = all_h[j]

print(f" best accuracy {best_accuracy:.4f}  k={best_k} h={best_h}")
```
